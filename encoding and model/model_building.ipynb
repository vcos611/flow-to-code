```python
def find_max_len(encodings):
    max_len = 0
    for key in encodings.keys():
        text = encodings[key]
        if len(text.split()) > max_len:
            #print(len(text.split()))
            max_len = len(text.split())
    return max_len

def perform_padding(encodings):
    max_len = find_max_len(encodings) + 1
    print("Maximum length is", max_len)
    for key in encodings.keys():
        text = encodings[key]
        sen_len = len(list(text.split()))
        if sen_len < max_len:
            add = ' <eos> ' * (max_len - sen_len)
            text += add
        encodings[key] = text
    #print(encodings)
    return encodings

            
def perform_preprocessing(encodings):
    preprocessed_dict = {}
    for key in encodings.keys():
        #print(key)
        preprocessed_text = ''
        for text, shape in encodings[key]:
            text_list= []
            new_text = ''
            for char in text:
                if char == '(' or char == ')' or char == ',' or char == '[' or char == ']':
                    new_text += ' '+char+' '
                elif char == "'":
                    pass
                else:
                    new_text += char

            combined_text = '{'
            for symbol in new_text.split():
                combined_text += ' ' + symbol + ' '
            combined_text += ' # '
            combined_text += ' ' + str(shape) + ' '
            combined_text += '}'
            #print(combined_text)
            preprocessed_text += combined_text + ' $ '
        #preprocessed_text = preprocessed_text[:-3]
        #print(preprocessed_text)
        preprocessed_dict[key] = preprocessed_text
        #print('---------------------------------------------------------------------')
    #print(preprocessed_dict)
    preprocessed_dict = perform_padding(preprocessed_dict)
    #print(preprocessed_dict)
    return preprocessed_dict


import pickle

def preprocess(input_filename, output_filename):
    fh = open(input_filename, 'rb')
    encodings = pickle.load(fh)
    fh.close()
    preprocessed_dict = perform_preprocessing(encodings)
    #print(preprocessed_dict)
    fh = open(output_filename, 'wb')
    pickle.dump(preprocessed_dict, fh)
    fh.close()

preprocess('encodings', 'preprocessed/input')
```

```python
import numpy as np

num_samples = 203

def build_vocabulary(encodings):
    vocab_list = ['<eos>']
    vocab_dict_word2idx = {}
    vocab_dict_idx2word = {}
    for key in encodings.keys():
        for word in encodings[key].split():
            if word not in vocab_list:
                vocab_list.append(word)
    #print(len(vocab_list))
    for idx, word in enumerate(vocab_list):
        vocab_dict_word2idx[word] = idx
        vocab_dict_idx2word[idx] = word
        
    #print(vocab_dict)
    return vocab_list, vocab_dict_word2idx, vocab_dict_idx2word

def get_one_hot_encodings(data, vocab, word2idx):
    vector = {}
    for key in data.keys():
        embedding = np.array([], dtype='int32')
        for word in data[key].split():
            embedding = np.append(embedding, word2idx[word])
        np.append(vector, embedding)
        vector[key] = embedding
    return vector

def convert_to_numpy(data, input_output):
    input_vector = np.array([], dtype='int32')
    vector_length = 0
    for i in range(1, num_samples+1):
        if input_output == 'input':
            key = 'FLOW__PID_'+str(i)+'.png'
        if input_output == 'output':
            key = 'CODE_PID_'+str(i)+'.py'
        vector = data[key]
        vector_length = len(vector)
        input_vector = np.append(input_vector, vector)

    input_vector = input_vector.reshape(num_samples, vector_length)
    #print(input_vector.shape)
    return input_vector


def get_input():
    fh = open('preprocessed/input', 'rb')
    inputs = pickle.load(fh)
    fh.close()

    input_vocab, input_word2idx, input_idx2word = build_vocabulary(inputs)
    input_one_hot_encoding_dict = get_one_hot_encodings(inputs, input_vocab, input_word2idx)

    input_data = convert_to_numpy(input_one_hot_encoding_dict, 'input')
    
    return input_data, input_vocab, input_word2idx, input_idx2word

def get_output():
    fh = open('preprocessed_code', 'rb')
    outputs = pickle.load(fh)
    fh.close()
    output_vocab, output_word2idx, output_idx2word = build_vocabulary(outputs)
    output_one_hot_encoding_dict = get_one_hot_encodings(outputs, output_vocab, output_word2idx)
    output_data = convert_to_numpy(output_one_hot_encoding_dict, 'output')
    return output_data, output_vocab, output_word2idx, output_idx2word




#get_input()
#get_output()
```

```python
from json import encoder
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Embedding, TimeDistributed

def seq2seq_model(input_seq_len, output_seq_len, input_vocab, output_vocab, HIDDEN_DIM=200):

    encoder_inputs = Input(shape = (None, input_seq_len, len(input_vocab)))
    #encoder_embedding = Embedding(input_dim = len(input_vocab), output_dim=HIDDEN_DIM, input_length=input_seq_len)(encoder_inputs)
    encoder_LSTM = LSTM(HIDDEN_DIM, return_state=True)
    #encoder_outputs, state_h, state_c = encoder_LSTM(encoder_embedding)
    encoder_outputs, state_h, state_c = encoder_LSTM(encoder_inputs)

    encoder_states = [state_h, state_c]

    decoder_inputs = Input(shape = (None, output_seq_len, len(output_vocab)))
    #decoder_embedding = Embedding(input_dim = len(output_vocab), output_dim = 32, input_length=output_seq_len)(decoder_inputs)
    decoder_LSTM = LSTM(HIDDEN_DIM, return_state=True, return_sequences=True)
    #decoder_outputs, _, _ = decoder_LSTM(decoder_embedding, initial_state = encoder_states)
    decoder_outputs, _, _ = decoder_LSTM(decoder_inputs, initial_state = encoder_states)

    decoder_outputs = TimeDistributed(Dense(len(output_vocab), activation='softmax'))(decoder_outputs)
    model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

    print(model.summary())
    return model

def convert_to_one_hot(input_data, output_data, input_vocab, output_vocab):
    encoder_input_data = np.zeros((num_samples, input_data.shape[1], len(input_vocab)), dtype='float32')
    decoder_input_data = np.zeros((num_samples, output_data.shape[1], len(output_vocab)), dtype='float32')
    decoder_target_data = np.zeros((num_samples, output_data.shape[1], len(output_vocab)), dtype='float32')

    for i, array in enumerate(input_data):
        for t, idx in enumerate(array):
            encoder_input_data[i, t, idx] = 1.
    
    for i, array in enumerate(output_data):
        for t, idx in enumerate(array):
            decoder_input_data[i, t, idx] = 1.
            if t > 0:
                decoder_target_data[i, t-1, idx] = 1.

    return encoder_input_data, decoder_input_data, decoder_target_data




input_data, input_vocab, input_word2idx, input_idx2word = get_input()
output_data, output_vocab, output_word2idx, output_idx2word = get_output()
encoder_input_data, decoder_input_data, decoder_target_data = convert_to_one_hot(input_data, output_data, input_vocab, output_vocab)
input_seq_len = input_data.shape[1]
output_seq_len = output_data.shape[1]



model = seq2seq_model(input_seq_len, output_seq_len, input_vocab, output_vocab, 32)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit([encoder_input_data, decoder_input_data], decoder_target_data, batch_size=16, epochs=10, validation_split=0.2)
```
